# Beregningsorientert øving i Fysikk/kjemi
I denne øvinga skal vi se nærmere på de to vanligste modellene for luft-/væskefriksjon, tilsvarende laminær og turbulent strøm.

## Bakgrunnsteori
Når et legeme beveger seg gjennom en gass eller en væske, virker det en kraft fra mediet på legemet som motvirker bevegelsen. I dagliglivet kalles disse kreftene luft- eller væskemotstand, avhengig av mediet.

Det er to hovedtyper "motstand" eller friksjonskrefter som avhenger av hvordan mediet strømmer rundt legemet:

<ul>
    <li><b>Laminær strøm</b> (laminær = "lagvis"), der mediet strømmer over overflaten til legemet i "ryddige" strømlinjer som ikke krysser hverandre. Typisk for <b>lave</b> hastigheter.</li>
    <li><b>Turbulent strøm</b>, der mediet ikke klarer å "henge fast på" overflaten til legemet, og det resulterende strømningsmønsteret blir kaotisk med strømlinjer som krysser hverandre. Typisk for <b> høye </b> hastigheter.</li>
</ul>

### Reynoldstallet
Hvorvidt strømmen er laminær eller turbulent, indikeres av det såkalte Reynoldstallet $N_R$ for strømmen, som er gitt ved
<br><br>
$$N_R=\frac{\rho v D}{\eta},$$
der $\rho$ er mediets massetetthet, $D=2r$ er legemets diameter, $\eta$ er legemets dynamiske viskositet, og $v$ er farten til legemet gjennom mediet. Vi bruker følgende grenser for å skille de to typene strømning:

Laminær strøm: $N_R \leq 2000$
<br>
Turbulent strøm: $N_R>2000$

### Laminær strøm (Stokes' lov for væskemotstand)
For små, kuleformede legemer som beveger seg i en væske(f.eks. leirepartikler eller sandkorn i vann), er væskefriksjonen gitt av Stokes' lov:
$$F_D = 6 \pi r \eta \cdot v,$$

der $\eta$ er mediets dynamiske viskositet  ([$\eta$]=Pa s), og $r$ er legemets radius.

 
### Turbulent strøm (for luftmotstand)
For legemer større enn $\sim \textrm{ mm}$ som beveger seg gjenno mluft, er strømmen turbulent og luftmotstanden er gitt ved:
$$F_D = \frac{1}{2} \rho A C_d \cdot v^2,$$

der $\rho$ er massetettheten til mediet, $A$ er "frontarealet" som ligger inn mot strømmen, og $C_d$ er den såkalte "drag coefficient" som er et dimensjonsløst tall på hvor strømlinjeformet legemet er. Typisk er $0,1 \lesssim C_d \lesssim 2,3$, der lavere verdier tilsvarer mer strømlinjeformede legemer.

## Eulers metode
Eulers metode brukes til finne en tilnærmet, numerisk løsning av en differensiallikning. Metoden kan brukes direkte for 1. ordens differensiallikninger, men kan lett utvides til differensiallikninger av høyere orden.

Vi skal bruke Eulers metode til å løse differensiallikninger på formen

$$\frac{dv}{dt}=a(t),$$

over et tidsrom $[t_0,t_1]$, med startbetingelsen $v(t_0)=v_0$. Akselerasjonen $a(t)$ varierer på grunn av den fartsavhengige friksjonskraften $F_D$.

Eulers metode deler inn tidsintervallet i $N$ like deler, med steglengde/tidssteg lik $h$. På et tidspunkt $t_n$ er farten $v_n$, og akselerasjonen er $a(t_n)$. Over et lite tidssteg $h$ er akselerasjonen tilnærmet konstant, slik at farten $v_{n+1}$ ved tidspunkt $t_{n+1}=t_n + h$ er gitt ved bevegelseslikningen for konstant akselerasjon:

$$v_{n+1}=v_n + a(t_n)h.$$

Dette gjentas så for alle punktene $n \leq N$:
$$v_1 = v_0 + a(t_0)h \\
v_2 = v_1 + a(t_1)h \\
\vdots \\
v_{N+1}=v_N+a(t_N)h$$
osv., slik at finner en tilnærmet løsning for $v(t)$ for $t \in [t_0,t_1]$.

Ettersom vi forutsetter konstant akselerasjon i løpet av tidssteget $h$, kan vi bestemme posisjonen $y(t)$ ved bevegelseslikningen $y=y_0+\frac{1}{2}(v_0+v)\Delta t$, dvs.
$$y_{N+1}=y_N+\frac{1}{2}(v_N+v_{N+1})h. $$

# Ferdige funksjoner
I denne delen finner du to ferdige funksjoner ```laminar``` og ```turbulent```, som beregner $v(t)$ og $y(t)$ for hhv. laminær og turbulent strøm. Ta utgangspunkt i denne koden når du besvarer oppgave 2.

````python
# Importerer bibliotek
import numpy as np # NumPy for behandling av matriser og ulike matematiske operasjoner
import matplotlib.pyplot as plt # For graftegning

# Sett standard parametre for plotting av grafer
figparam = {'figure.figsize': (15, 7), 'axes.grid': False,
             'lines.markersize': 10, 'lines.linewidth': 2,
             'font.size': 15, 'mathtext.fontset': 'stix',
             'font.family': 'STIXGeneral', 'figure.dpi': 200}
plt.rcParams.update(figparam)

# Globale variable
g=9.81 #Tyngdeakselerasjonen i m/s^2
C_d=0.5 #Legemets "drag coefficient"
rho_l=1.2 #Massetetthet for luft i kg/m^3
rho_v=1000 #Massetetthet for vann i kg/m^3
eta=0.00089 #Dynamisk viskositet for vann ved 25 grader, i Pa s
````

````python
def laminar(m, r, v0, t0,t1,h):
       
    """Bruker Eulers metode med steglengde h fra t0 til t1 til å bestemme v(t) og y(t) for et legeme med gitt 
    startfart v0. Bruker Stokes' lov under forutsetning at strømmen er laminær, dvs. friksjonskraften er F_D = kv.

    Parametre:
            m: Legemets masse [kg]
            r: Legemets radius [m]
            v0: Startfart [m/s]
            t0:Starttidspunkt [s]
            t1:Sluttidspunkt [s]
            h: Tidssteg [s]
            
    Returnerer:
            t (array fra t0 til t1), v (array fra startfart v0 til sluttfart), y (array fra startposisjon y0=0 til 
            sluttposisjon)
    """
    
    k=6*np.pi*r*eta #Proporsjonalitetskontanten i Stokes' lov, F_D = kv
    N=int((t1-t0)/h) #N er antall tidssteg i simuleringen
    v=np.zeros(N+1) #Oppretter array for v(t), der alle v(t) settes lik 0 i utgangspunktet
    y=np.zeros(N+1) #Oppretter array for y(t), der alle y(t) settes lik 0 i utgangspunktet
    t = np.linspace(t0, t1, N + 1) #Oppretter array for tidsverdier, [t0,...,t1]
    v[0]=v0 #Legger inn startverdi for fart
    y[0]=0 #Startverdi for y
    for i in range(0,N): #Bruker Eulers metode med N stk. tidssteg lik h
        a=g-(k/m)*v[i] #Beregn akselerasjonen a for gitt fart
        v[i+1]=v[i]+a*h #Beregn v(t+h) med ett steg av Eulers metode
        y[i+1]=y[i]+0.5*(v[i]+v[i+1])*h #Beregn y(t+h) vha. bevegelseslikning for konstant aks.
    return t,v,y
````

````python
def turbulent(m,r,v0, t0,t1,h):
    
    """Bruker Eulers metode med steglengde h fra t0 til t1 til å bestemme v(t) og y(t) for et legeme med gitt 
    startfart v0. Bruker Stokes' lov under forutsetning at strømmen er turbulent, dvs. friksjonskraften er F_D = kv^2.

     Parametre:
            m: Legemets masse [kg]
            r: Legemets radius [m]
            v0: Startfart [m/s]
            t0:Starttidspunkt [s]
            t1:Sluttidspunkt [s]
            h: Tidssteg [s]
            
    Returnerer:
            t (array fra t0 til t1), v (array fra startfart v0 til sluttfart), y (array fra startposisjon y0=0 til 
            sluttposisjon)
    """
    A=np.pi*r**2 #Legemets frontareal, lik 1/4 av en kuleoverflate ved kuleformet legeme
    k=0.5*rho_l*A*C_d #Proporsjonalitetskonstanten i F_D=kv^2
    N=int((t1-t0)/h) #N er antall tidssteg i simuleringen
    v=np.zeros(N+1) ##Oppretter array for v(t), der alle v(t) settes lik 0 i utgangspunktet
    y=np.zeros(N+1) #Oppretter array for y(t), der alle y(t) settes lik 0 i utgangspunktet
    t = np.linspace(t0, t1, N + 1) #Oppretter array for tidsverdier, [t0,...,t1]
    v[0]=v0 #Legger inn startverdi for fart
    y[0]=0 #Startverdi for y
    for i in range(0,N): #Bruker Eulers metode med N stk. tidssteg lik h
        a=g-(k/m)*v[i]**2 #Beregn akselerasjonen a for gitt fart
        v[i+1]=v[i]+a*h #Beregn v(t+h) med ett steg av Eulers metode
        y[i+1]=y[i]+0.5*(v[i]+v[i+1])*h #Beregn y(t+h) vha. bevegelseslikning for konstant aks.
    return t,v,y
````

````python
#Eksempel på plotting av fartsgraf med matplotlib
(t,v,y)=laminar(1.0E-6,1.0E-4,0,0,5,0.1)
fartsgraf=plt.subplot()
fartsgraf.set_title("Fartsgraf for fallende legeme ved laminær strøm")
fartsgraf.set_xlabel(r"$t$ [s]")
fartsgraf.set_ylabel(r"$v$ [m/s]")
fartsgraf.plot(t,v)
plt.show()
````
